### 1. which establishments have a hygiene score equal to 20?

```python
# Find establishments with a hygiene score equal to 20
query = {"scores.Hygiene": 20}
results = establishments.find(query)

# Use a loop to count the documents in the result
count = 0
for _ in results:
    count += 1

print("Number of documents with hygiene score equal to 20:", count)

# Display the first document in the results using pprint
if count > 0:
    # You'll need to create a new cursor since the previous one is exhausted
    first_document = establishments.find(query)[0]
    pprint(first_document)
else:
    print("No establishments with a hygiene score of 20 found.")
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "AddressLine4": "London",
    "RatingValue": {"$gte": "4"}
}
# Find establishments that match the query
#results = establishments.find(query)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print("Number of establishments in London with a rating of 4 or higher:", count)

# Display the first document in the results using pprint
if count > 0:
    first_document = establishments.find_one(query)
    pprint(first_document)
else:
    print("No establishments in London with a rating of 4 or higher found.")
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query)
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the name of the restaurant to search for
restaurant_name = "Penang Flavours"

# Define the query to find the restaurant by name
query = {"BusinessName": restaurant_name}

# Use find_one to search for the restaurant
restaurant = establishments.find_one(query)

# Check if the restaurant was found
if restaurant:
    print(f"{restaurant_name} is in the dataset.")
    # Optionally, you can print the details of the restaurant
    print("Restaurant Details:")
    pprint(restaurant)
else:
    print(f"{restaurant_name} is not in the dataset.")
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = float("0.01")
latitude = float("51.49014200") 
longitude = float("0.08384000")

query = {
    "geocode.latitude": {
        "$gte": latitude - degree_search,
        "$lte": latitude + degree_search
    },
    "geocode.longitude": {
        "$gte": longitude - degree_search,
        "$lte": longitude + degree_search
    },
    "RatingValue": 5
}

sort = [("scores.Hygiene", 1)]

# Use find method to retrieve the top 5 establishments that match the criteria
results = establishments.find(query).sort(sort).limit(5)

# Print the results
print("Top 5 establishments with RatingValue 5, sorted by lowest hygiene score:")
for idx, establishment in enumerate(results, start=1):
    print(f"{idx}. Name: {establishment['BusinessName']}, Hygiene Score: {establishment['scores:','Hygiene']}")
```

```python
# Convert result to Pandas DataFrame
# Create an empty list to store the documents
document_list = []

# Iterate through the MongoDB query results and append each document to the list
for establishment in results:
    document_list.append(establishment)

# Create a Pandas DataFrame from the list of documents
df = pd.DataFrame(document_list)

# Display the Pandas DataFrame
print("Pandas DataFrame:")
print(df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {"count": -1}
    }
]
#Execute the aggregation pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in the result:", len(result))
# Print the first 10 results
print("First 10 results:")
pprint(result[:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)
# Display the number of rows in the DataFrame
num_rows = len(df)
print(f"Number of rows in the DataFrame: {num_rows}")
# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
df.head(10)
```
